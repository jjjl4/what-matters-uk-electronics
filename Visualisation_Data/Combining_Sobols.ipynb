```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd



%matplotlib inline
```

```python
washingmachine = pd.read_excel('../Output_Data/WashingMachine_ST.xlsx')
washingmachine_sobol = washingmachine.iloc[:,0]

fridge = pd.read_excel('../Output_Data/FridgeFreezer_ST.xlsx')
fridge_sobol = fridge.iloc[:,0]

tv = pd.read_excel('../Output_Data/TV_ST.xlsx')
tv_sobol = tv.iloc[:,0]

monitor = pd.read_excel('../Output_Data/Monitor_ST.xlsx')
monitor_sobol = monitor.iloc[:,0]

desktop = pd.read_excel('../Output_Data/Desktop_ST.xlsx')
desktop_sobol = desktop.iloc[:,0]

laptop = pd.read_excel('../Output_Data/Laptop_ST.xlsx')
laptop_sobol = laptop.iloc[:,0]

smartphone = pd.read_excel('../Output_Data/Smartphone_ST.xlsx')
smartphone_sobol = smartphone.iloc[:,0]


labels_data = pd.read_excel('../Output_Data/Sobol_Labels.xlsx')
labels = labels_data.iloc[:,0]


```

```python
def remove_alpha(sobol_data):
    del sobol_data[1]
    del sobol_data[3]
    return sobol_data
```

```python
washingmachine_sobol = remove_alpha(washingmachine_sobol)
fridge_sobol = remove_alpha(fridge_sobol)
tv_sobol = remove_alpha(tv_sobol)
monitor_sobol = remove_alpha(monitor_sobol)
desktop_sobol = remove_alpha(desktop_sobol)
laptop_sobol = remove_alpha(laptop_sobol)
smartphone_sobol = remove_alpha(smartphone_sobol)
```

```python
plt.style.use('bmh')

edited_labels = ['Devices in Service', 'Primary Lifetime', 'Secondary Lifetime', 'Repair Rate', 
                 'Size of Device','Power Consumption', 'Usage Behaviour']
width = 0.6
y = np.arange(len(edited_labels))


fig = plt.figure(figsize=(18,10))
gs = fig.add_gridspec(1,7, wspace=0.15)
ax= gs.subplots(sharex=True, sharey=True)

fig.suptitle('Sobol Indices for the Impact of Model Variables on Final Energy Demand in 2050', fontsize=20)



ax[0].barh(y, washingmachine_sobol, width, label='Washing Machine', color='#006666')
ax[0].set_title('Washing Machine')

ax[1].barh(y , fridge_sobol, width, label='Fridge', color='#008080')
ax[1].set_title('Fridge Freezer')


ax[2].barh(y , tv_sobol, width, label='TV', color= '#011f4b')
ax[2].set_title('TV')

ax[3].barh(y , monitor_sobol, width, label='Monitor', color= '#03396c')
ax[3].set_title('Monitor')

ax[4].barh(y , desktop_sobol, width, label='Desktop', color= '#005b96')
ax[4].set_title('Desktop')

ax[5].barh(y , laptop_sobol, width, label='Laptop', color= '#6497b1')
ax[5].set_title('Laptop')

ax[6].barh(y , smartphone_sobol, width, label='Smartphone', color= '#b3cde0')
ax[6].set_title('Smartphone')


ax[0].set_yticks(y)
ax[0].set_yticklabels(edited_labels, fontsize=14)


plt.xlim([0,1])
plt.xticks([0,0.5,1])


plt.savefig('Figures/Combined_Sobols.pdf', dpi=400, bbox_inches='tight')
plt.show()
```
